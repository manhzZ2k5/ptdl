```python
import pandas as pd

# 1. Đọc file gốc
df_raw = pd.read_csv('../media_all_channels.csv')

# 2. Danh sách các cột cần DROP (Xóa)
# Bao gồm mọi thứ không có sẵn trước khi chiến dịch bắt đầu
columns_to_drop = [
    'impressions',      # Chưa chạy nên chưa biết có bao nhiêu hiển thị
    'clicks',           # Chưa chạy nên chưa biết có bao nhiêu click
    'conversions',      # Kết quả mua hàng
    'cpc', 'cpa',       # Chi phí phái sinh
    'ctr',              # Tỷ lệ click
    'conversion_rate',  # Tỷ lệ chuyển đổi
    'roas', 'roi',      # Chỉ số hoàn vốn
    'profit_margin'     # Biên lợi nhuận
]

# 3. Tạo DataFrame mới
df_model = df_raw.drop(columns=columns_to_drop)

# 4. Kiểm tra lại: Tập dữ liệu bây giờ sẽ cực kỳ tinh gọn!
print("Các cột còn lại để đưa vào mô hình:")
print(df_model.columns.tolist())

# Xem trước 3 dòng
display(df_model.head(3))

# 5. Xuất ra file CSV mới
df_model.to_csv('../media_for_forecasting.csv', index=False)
print("\nĐã tạo thành công file: media_for_forecasting.csv")
```
