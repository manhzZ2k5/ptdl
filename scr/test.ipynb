```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
df=pd.read_csv('../media_all_channels.csv')
channels = df['channel'].unique()

# To khung li biu  (2 dng, 3 ct)
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))
axes = axes.flatten() # Lm phng mng axes  d chy vng lp

# Chy vng lp qua tng knh  v
for i, channel in enumerate(channels):
    # Lc d liu ca knh 
    subset = df[df['channel'] == channel]
    
    # --- BC 1: V CC CHM (SCATTER) BNG PANDAS ---
    # ax=axes[i] ngha l v vo  th i trong li
    subset.plot.scatter(x='cost', y='revenue', ax=axes[i], 
                        color='skyblue', alpha=0.6, s=30, label='D liu thc')
    
    # --- BC 2: V NG XU HNG (TRENDLINE) ---
    # Pandas khng t v c, phi dng Numpy tnh phng trnh ng thng y = ax + b
    if len(subset) > 1: # Ch v nu c nhiu hn 1 im d liu
        z = np.polyfit(subset['cost'], subset['revenue'], 1) # Tm h s a, b
        p = np.poly1d(z) # To phng trnh
        
        # V ng thng  ln biu 
        axes[i].plot(subset['cost'], p(subset['cost']), "r--", linewidth=2, label='Xu hng')

    # Trang tr
    axes[i].set_title(f'Knh: {channel}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel('Chi ph (Cost)')
    axes[i].set_ylabel('Doanh thu (Revenue)')
    axes[i].legend()

# Cn chnh khong cch cho p
plt.tight_layout()
plt.show()
```
